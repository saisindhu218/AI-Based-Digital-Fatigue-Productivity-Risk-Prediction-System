```python
import pandas as pd

df = pd.read_csv('../datasets/digital_usage_data.csv')
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(df['screen_time'], df['fatigue_score'])
plt.xlabel('Screen Time (hours)')
plt.ylabel('Fatigue Score')
plt.title('Screen Time vs Fatigue Score')
plt.show()
```

```python
df['fatigue_level'].value_counts().plot(kind='bar')
plt.title('Fatigue Level Distribution')
plt.show()
```

### Feature Understanding

- **screen_time**: Higher daily screen exposure increases cognitive fatigue.
- **avg_session**: Longer continuous sessions indicate reduced breaks, leading to fatigue.
- **breaks**: More frequent breaks reduce fatigue accumulation.
- **night_ratio**: Late-night usage disrupts circadian rhythm and increases fatigue.
- **productive_ratio**: Higher productive app usage is associated with better focus and lower fatigue.
- **fatigue_score**: Represents cumulative digital fatigue on a 0100 scale.
- **fatigue_level**: Categorized fatigue risk (Low, Medium, High).
- **productivity_loss**: Estimated weekly productivity loss in hours due to fatigue.
